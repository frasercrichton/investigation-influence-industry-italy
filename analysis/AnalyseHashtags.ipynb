# TikTok Hashtags

## Hashtags in Italian

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import logging
import matplotlib.pyplot as plt
import os
import numpy as np
import sys

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
stopwords = set(STOPWORDS)

logging.getLogger("matplotlib.font_manager").setLevel(logging.ERROR)
np.set_printoptions(threshold=sys.maxsize)

processed_dir =  '../data/processed/'
os.makedirs(processed_dir + 'images', exist_ok = True)

hashtags_dir = processed_dir + 'hashtags'
hashtags_it_df = pd.read_csv(hashtags_dir + '/it_hashtags.csv')
```

```python
f = plt.figure()
f.set_figwidth(44)
f.set_figheight(50)
plt.barh(hashtags_it_df['textExtra'], hashtags_it_df['count'], label = "Hashtag Counts", color='red')
font = {'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)
plt.legend()
plt.savefig(processed_dir + 'images/it-hashtags.png')
plt.show()
```

```python
hashtags_dict = dict(zip(hashtags_it_df['textExtra'].tolist(), hashtags_it_df['count'].tolist()))
stopwords = set(STOPWORDS)
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

# max_wordsnumber (default=200)
wordcloud = WordCloud(background_color="white", mask=mask).generate_from_frequencies(hashtags_dict)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig(processed_dir + 'images/it-hashtags-word-cloud.png')
plt.show()
```

## Hashtag Counts - English

```python
hashtags_en_df = pd.read_csv(hashtags_dir + '/en_hashtags.csv')
hashtags_en_df = hashtags_en_df.sort_values(by=['count'], ascending=True)
f = plt.figure()
f.set_figwidth(44)
f.set_figheight(50)
plt.barh(hashtags_en_df['textExtra_en'], hashtags_en_df['count'], label = "Hashtag Counts", color='red')
font = {'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)
plt.legend()
plt.savefig(processed_dir + 'images/en-hashtags-barchart.png')
plt.show()
```

```python
en_hashtags_dict = dict(zip(hashtags_en_df['textExtra_en'].tolist(), hashtags_en_df['count'].tolist()))
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)
wordcloud = WordCloud(background_color="white", mask=mask).generate_from_frequencies(en_hashtags_dict)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig(processed_dir + 'images/it-hashtags-word-cloud.png')
plt.show()
```
